```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
tips.head()
```

```python
# BAR PLOT: aggregates categorical data from a specific function

# x is categorical
# y is numeric

# we submit the dataframe to data

# the estimator param is a function that determines categorical bin (?)
# not sure how estimator works... come back to that later

import numpy as np

sns.barplot(x="sex", y="total_bill", data=tips, estimator=np.mean)

plt.show()
plt.close()
```

```python
# COUNTPLOT: barplot, but only counts ocurrences of qualitative data
# that's why it only needs an x param!

sns.countplot(x="sex", data=tips)

plt.show()
plt.close()
```

```python
# boxplot func
# it creates a boxplot for numerical values divided into categories

# points outside whiskers are outliers!
# 'hue' does same thing as before; it "colors" by category

sns.boxplot(x="day", y="total_bill", data=tips, hue="smoker")

plt.show()
plt.close()
```

```python
# violin plot is like a boxplot, but it shows kernel density est. in its shape

sns.violinplot(x="day", y="total_bill", data=tips, hue="sex", split=True)

# remember u can also use hue here!
# if u add split it doesnt create 2 diff charts but makes it so each half pertains to diff hue
```

```python
# STRIP PLOT: scatterplot where 1 var is categorical; shows ocurrences of numerical data for that category

# jutter arg adds sideways noise to distinguish stacked points

sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
```

```python
# swarm plot is strip plot without stacked points
# it also shows distribution!
# combines stripplot with violin plot

# NOTE: don't use these too often; they aren't very well known and people might have trouble reading them

sns.violinplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips, color="black") # calling both in sequence plots them on same figure!
```
