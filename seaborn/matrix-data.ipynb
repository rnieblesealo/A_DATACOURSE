```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# heatmaps are useful to show matrix data

# a dataframe is NOT a numeric matrix! we need to convert it into one
# we can use pivoting or correlation

# here we use correlation
# it correlates the numeric values and creates a matrix from the output

tc = tips.corr(numeric_only=True)
tc
```

```python
# heatmaps replace matrix values for squares to indicate their magnitude!
sns.heatmap(tc, annot=True, cmap="coolwarm")

plt.show()
plt.close()
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
# reshape table
fp = flights.pivot_table(index="month", columns="year", values="passengers", observed=False)
fp
```

```python
# add sep with linewidths and linecolor!

sns.heatmap(fp, cmap="magma", linewidths=0.1, linecolor="black")

plt.show()
plt.close()
```

```python
# matrices can also be described as clustermaps

# they help us see clusters of values
# data is rearranged to represent clusters

sns.clustermap(fp, cmap="coolwarm", standard_scale=1) 

# coolwarm helps see pattern!
# standardscale normalizes data to clarify even further!

plt.show()
plt.close()
```
