```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
plt = sns.lmplot(x='total_bill', y='tip', data=tips,
    hue='sex',
    markers=['o', 'v'], # we can marker each diff category in the hue!
    # we can pass matplotlib args directly:
    scatter_kws={
        's': 100
    }
)
```

```python
# we can separate hue values by passing col arg
# we can row it by passing another qual as row=
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time', aspect=0.8)
```
