```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips') # builtin datastet
tips.head()
```

```python
import matplotlib.pyplot as plt

displot = sns.displot(tips['total_bill'], kde=True, bins=10) # distribution chart; quantity histogram

plt.show() # need to run with matplotlib plots; seaborn hooks up to mpl backend!
plt.close()
```

```python
# joint plot; combines 2 distributions mapping 2 diff variables
# kind param. determines how dists are compared!

# 'hex' to style dist as hex
# 'reg' to show linear regression line
# 'kde' to show squiggly distribution

sns.jointplot(x='total_bill',y='tip',data=tips, kind='kde')

plt.show()
plt.close()
```

```python
# pair plot: plots pairwise relationships for all dataframe
# is like a jointplot for every possible combination of 2 numerical vars
# woah :0

# 'hue' colors numeric datapoints by considering a non-numeric category
# this is so sick!

sns.pairplot(tips, 
    hue="sex",
    palette="coolwarm" # theme
)

plt.show()
plt.close()
```

```python
# histogram works with bins
# a bin counts how many datapoints fall inside of it to then display its bar in accordance with that count

# a RUG PLOT is the same but with no bins; just draw a dash for every point in the distline

sns.rugplot(tips['total_bill'])

plt.show()
plt.close()
```

```python
sns.displot(tips['total_bill'])

plt.show()
plt.close()
```

```python
# a KDE PLOT is a KERNEL DENSITY ESTIMATION plot
# like a regression curve for a probabilistic distribution
# it estimates the probability at a point based on the given data

# is like a sum of all the normal dists of every rug point!

# we can add it to our histogram vis with 'kde=True'

sns.displot(tips['total_bill'], kde=True)

plt.show()
plt.close()
```
