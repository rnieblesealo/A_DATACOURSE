```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
# sns.pairplot(iris) # pairplots between all variables
pg = sns.PairGrid(iris) # pairgrid is empty pairplot

pg.map(plt.scatter) # map "fills" grid with scatterplots
pg.map_diag(sns.displot) # we can change graph type diagonal, upper, lower...
# pg.map_upper()
# pg.map_lower()

plt.show()
plt.close()
```
