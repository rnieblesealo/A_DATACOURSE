```python
# matplotlib.org/gallery to get chart references

import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11) # x axis; equivalently spaced
y = x ** 2 # y axis; the "function"
```

```python
# functional and objectorient methods for making plots
```

```python
# functional method
plt.plot(x, y) # jupyter issue: printing object but not img itself...

# adding labels and title
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("f(x) = x^2")

plt.show() # ...calling this is the soln
```

```python
# multiple plots on same canvas
plt.subplot(1, 2, 1) # 1: 1 row, 2: 2 cols, 1: my index is 1
plt.plot(x, y, 'r') # 'r' for red

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')

plt.show()
```

```python
# oop method; instructor says it's better
fig = plt.figure() # figure object is the canvas

# add axes to canvas, axes determine a subplot I THINK
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, w, h

axes.set_xlabel('x')
axes.set_ylabel('f(x)')
axes.set_title('f(x) = x^2')

# plot!
axes.plot(x, y)

fig
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.set_title("Big plot!")
axes1.plot(y, x, 'b')

# bottom corner
axes2 = fig.add_axes([0, 0, 0.4, 0.3])
axes2.set_title("Baby plot")
axes2.plot(x, y, 'r')

fig
```

```python
# oop subplots
fig, axes = plt.subplots(nrows=1, ncols=2) # this returns a tuple, hence weird assignment

# axes is a list containing plots so we can treat it like a list!
for i, curr_ax in enumerate(axes):
    curr_ax.set_title(f"Plot {i + 1}")
    curr_ax.plot(x, y)

plt.tight_layout()

fig
```

```python
# we can specify size of figure
fig = plt.figure(figsize=(5, 2)) # in inches!
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y)

fig
```

```python
# can also do it with multiple figs
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 2))

axes[0].plot(x, y)
axes[1].plot(y, x)

fig.tight_layout()

fig
```

```python
# saving fig to file
# specified extension will be the one used; matplot figures it out for us
fig.savefig("myfigure.png")
```

```python
# multiple plots on same graph!
# just plot multiple times on same axis :)

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='r', label="Normal")
ax.plot(y, x, color='g', label="Inverse")
```

```python
# adding a legend for the multiple plots!
# need to give sep plots labels

ax.legend(loc=0) # specify location loc with codes; check doc page for meaning of diff ones; 0 is auto

# can also do custom coord using percent tuple
ax.legend(loc=(0.1, 0.1))

fig
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

# set line color (basics are named, can use rgb hex too)
# line width (can set as linewidth= or lw=), 
# line alpha (0-1)
# linestyle (can also use as ls=) (-- for dashed, : for dotted, - for normal)
# markers and their size (markers are placed at the datapoints we supplied)
ax.plot(x, y, 
        color="#FF8C00", 
        lw=1, 
        alpha=0.6, 
        linestyle="-",
        marker="o", # "o" for circle "+" for cross
        markersize=10,
        markerfacecolor="#000000", # fill of marker
        markeredgewidth=4, # marker stroke width
        markeredgecolor="red" # marker stroke color
) 

# KEY TAKEAWAY: there is a lot of customization here, check out docs!

# can set low and high bound of x
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])

fig
```
